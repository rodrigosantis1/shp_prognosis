```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

### Leitura dos sensores

```python
#Constants
UG=1
FRQ='5min'
N = 10
```

```python
df = pd.read_csv("../data/AdoPopinhaki_UG{}.csv".format(UG), parse_dates=['E3TimeStamp'])
#df = df.loc[df['Gerador - Potência Aparente Média'] > 1000]
```

```python
df.head()
```

```python
df2 = df.set_index('E3TimeStamp').resample(FRQ).mean()
```

```python
ax = df2['Vibração Mancal Combinado Radial Vertical'].plot(figsize=(14,3))
ax.set_ylabel('Vibration')
```

```python
df3 = df2.dropna()
```

```python
plt.figure(figsize=(14,3))
plt.plot(df3['Vibração Mancal Combinado Radial Vertical'].values)
```

```python
df_list = []
for ug in range(1,5):
    df = pd.read_csv("../data/AdoPopinhaki_UG{}.csv".format(ug), 
                     parse_dates=['E3TimeStamp'], index_col=['E3TimeStamp']).resample(FRQ).mean().dropna()
    vibr_cols = [col for col in df.columns if 'Vibração' in col]
    
    # Draw random
    np.random.seed(ug*42)
    cut_points = np.sort(np.random.randint(len(df)//N, len(df), size=N))
    
    for _, cp in enumerate(cut_points):
        new_cols = ['UG{}X{}P{}'.format(ug,i,_) for i in range(len(vibr_cols))]

        ts = df[vibr_cols].iloc[:cp]
        ts.columns = new_cols
        ts = ts.transpose()
        df_list.append(ts)
```

```python
df2 = pd.concat(df_list)
df2
```

```python
df2.to_csv('../data/univariate.csv')
```
